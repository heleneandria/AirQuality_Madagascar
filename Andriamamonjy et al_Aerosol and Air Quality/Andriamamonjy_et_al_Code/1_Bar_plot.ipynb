```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import dates as mpl_dates
%matplotlib inline 
sns.set(color_codes=True)
```

```python
plt.style.use('default')
```

```python
data1 = pd.read_excel('Ambatobe.xlsx')
data2 = pd.read_excel('Ampandrianomby.xlsx')
data3 = pd.read_excel('Andraharo.xlsx')
data4 = pd.read_excel('Ambohidahy.xlsx')
data5 = pd.read_excel('Antsakaviro.xlsx')
data6 = pd.read_excel('Soanierana.xlsx')
data7 = pd.read_excel('Miandrarivo.xlsx')
```

```python
data = pd.concat([data1, data2, data3, data4, data5, data6, data7])
```

```python
data.head(24)
```

```python

def plot_data(ax, data, title, label):
    color_list = ["green", "yellow", "orange", "red", "purple", "#BA0001"]
    ax = data.plot(ax=ax, x='Month', y=["Good", "Moderate", "Unh. For S.G.", "Unhealthy", "V. Unhealthy", "Hazardous"], 
                   kind='bar', stacked=True, mark_right=False, color=color_list, legend=False)

    ax.spines["top"].set_color('black')
    ax.spines["bottom"].set_color('black')
    ax.spines["left"].set_color('black')
    ax.spines["right"].set_color('black')
    ax.tick_params(bottom=True, top=False, left=True, right=False)
    
    ax.set_xlabel('', fontsize=14)  # Pas de label spcifique pour l'axe des x
    ax.tick_params(axis='x', labelsize=18, rotation=90)  # Rotation des labels de l'axe x
    ax.tick_params(axis='y', labelsize=18)
    ax.set_title(title, fontsize=18)
    ax.text(0.05, 0.95, label, transform=ax.transAxes, fontsize=25, fontweight='bold', ha='right', va='top')


n = 7  
ncols = 2
nrows = int(np.ceil(n / ncols))

fig, axs = plt.subplots(nrows, ncols, figsize=(20, 6 * nrows))

plot_data(axs[0, 0], data1, 'Case Ambatobe', 'A')
plot_data(axs[0, 1], data2, 'Case Ampandrianomby', 'B')
plot_data(axs[1, 0], data3, 'Case Andraharo', 'C')
plot_data(axs[1, 1], data4, 'Case Ambohidahy', 'D')
plot_data(axs[2, 0], data5, 'Case Antsakaviro', 'E')
plot_data(axs[2, 1], data6, 'Case Soanierana', 'F')
plot_data(axs[3, 0], data7, 'Case Miandrarivo', 'G')


fig.text(0.04, 0.5, 'Percentage of the day (%)', va='center', rotation='vertical', fontsize=35)

# Dsactiver le dernier subplot (bas  droite)
axs[3, 1].axis('off')

# lgende globale
handles, labels = axs[0, 0].get_legend_handles_labels()
axs[3, 1].legend(handles, labels, loc='best', fontsize=30)


plt.tight_layout(rect=[0.05, 0, 1, 1])  # Ajustement pour viter que le label y se chevauche

#plt.savefig('Classification_each_PA.png')

plt.show()
```
